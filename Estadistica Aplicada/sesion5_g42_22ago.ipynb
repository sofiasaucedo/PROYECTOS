```python
## Librerias 

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# datos 
personas = ['Jorge','Raul','Ana','Lissa']
edades = [27,16,21,30]

dic1= {'Municipio':['Escobedo','Apodaca','Guadalupe','Santiago'],'Nacimientos':[123,345,278,321],'Muertes':[23,17,56,32],'Matrimonios':[76,45,89,12]}

datos_mun=pd.DataFrame(dic1)
```

```python
datos_mun
```

```python
personas, edades
```

```python
# Grafica

plt.bar(personas,edades)
plt.title('Edades')
plt.xlabel('Nombres')
plt.ylabel('Edad en a√±os')
plt.show()
```

```python
plt.bar(datos_mun['Municipio'],datos_mun['Matrimonios'])
plt.title('Matrimonios en N.L')
plt.xlabel('Municipio')
plt.ylabel('Numero de matrimonios')
plt.show()
```

```python
fig, (graf1,graf2)= plt.subplots(2,1)
```

```python
#Pizarron 
fig, (graf1,graf2)= plt.subplots(2,1)

## Grafica 1 (edades)
graf1.bar(personas,edades)

## Grafica 2 (matrominios)
graf2.bar(datos_mun['Municipio'],datos_mun['Matrimonios'])
plt.show()
```
