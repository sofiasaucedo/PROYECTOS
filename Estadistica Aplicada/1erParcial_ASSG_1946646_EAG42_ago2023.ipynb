### Ejercicio 1 

```python
# Importamos base de datos de peliculas

datos_peliculas = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/peliculas.csv')
```

```python
datos_peliculas.head()
```

### a)

```python
# Cantidad de peliculas 

peliculas_cantidad = datos_peliculas['title'].count()

print('La cantidad de peliculas que hay son :',peliculas_cantidad)
```

```python
# Cantidad de generos  

generos_agrupados = datos_peliculas.groupby('category').agg('count')
generos = generos_agrupados.index

generos_cantidad= generos.value_counts().sum()

print('La cantidad de generos que hay son :',generos_cantidad)
```

```python
# Duracion promedio de las peliculas 

duracion_promedio = datos_peliculas['length'].mean()

print('Duración en promedio de las peliculas :',duracion_promedio)
```

### b)

```python

peliculas_dormir= datos_peliculas[(datos_peliculas['length']>120) & (datos_peliculas['category']=='Documentary')]

peliculas_dormir
```

### c)

```python
# Cantidad peliculas por genero

peliculas_por_genero = datos_peliculas.groupby('category')['title'].agg('count')
peliculas_por_genero
```

### d)

```python
# Calcular la duración promedio de las películas de cada género

duracion_por_genero = datos_peliculas.groupby('category')['length'].agg('mean')
duracion_por_genero
```

### e)

```python
# Nueva columna 

condiciones = [datos_peliculas['length']< 90 , datos_peliculas['length']<=150,datos_peliculas['length']>150]

valores = ['Corta','Normal','Larga']

datos_peliculas['Duracion']= np.select(condiciones,valores,default='Missing')

datos_peliculas
```

### Ejercicio 2

```python
# Importamos base de datos de la poblacion Asgard 

datos_Asgard = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/poblacion_Asgard.csv')
```

```python
type(datos_Asgard)
```

```python
datos_Asgard.head()
```

### a)

```python
#Ordena la base alfabéticamente por estado y por municipio

datos_Asgard.sort_values(['Estado','Municipio'])
```

### b)

```python
# Cantidad de estados en Asgard 

estados_agrupados = datos_Asgard.groupby('Estado').agg('count')
estados= estados_agrupados.index

estados_cantidad= estados.value_counts().sum()


print('La cantidad de estados que hay son :', estados_cantidad)
```

```python
# Cantidad de municipios que hay en cada estado

municipios_por_Estado = datos_Asgard.groupby('Estado')['Municipio'].agg('count')

municipios_por_Estado 
```

```python
# Municipios en total 

total_municipios= datos_Asgard['Municipio'].count()
print('La cantidad de municipios que hay son :', total_municipios)
```

### c)

```python
# Población que tiene el país en total

poblacion_total= datos_Asgard['Población'].sum()
print('La cantidad de la poblacion que hay en Asgard :', poblacion_total)
```

```python
# Población que tiene cada estado

estados_agrupados = datos_Asgard.groupby('Estado').agg('sum')
estados_agrupados
```

### d)

```python
#Calcula la mediana de la población de los municipios, y clasifícalos, 
#aquellos por arriba de la mediana deben llamarse “muy poblados” 
#y aquellos debajo de la mediana “poco poblados”

mun_med =datos_Asgard.groupby('Municipio')['Población'].sum()
mun_med=mun_med.median()
print('La mediana de la población de los municipios es:',mun_med)
```

```python
datos_Asgard['Tamaño población']=np.where(datos_Asgard['Población']>mun_med,'muy poblados','poco poblados')
datos_Asgard
```

### e)

```python
# Encuentra los 5 municipios mas poblados y los cinco municipios menos poblados del país

#los 5 municipios mas poblados

maspoblados=datos_Asgard.sort_values('Población', ascending = False)
maspoblados.head()
maspoblados[['Municipio','Población']].head()
```

```python
# los cinco municipios menos poblados del país
menospoblados=datos_Asgard.sort_values('Población')
menospoblados.head()
menospoblados[['Municipio','Población']].head()
```

### f)

```python
# El municipio más poblado de cada estado
estado_maspoblado=datos_Asgard.groupby('Estado')['Población'].idxmax()
datos_Asgard.loc[estado_maspoblado]
```

```python
# Genera un grafico de barras de los 15 municipios más poblados de Asgard

municipios_mas=datos_Asgard.sort_values('Población',ascending=False).head(15)
plt.bar(municipios_mas['Municipio'],municipios_mas['Población'])
plt.title('Los 15 Municipios más Poblados de Asgard')
plt.xlabel('Municipio')
plt.ylabel('Hanitantes')
plt.ticklabel_format(style="plain", axis="y")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Genera un grafico de barras de Cuantas peliculas hay de cada género

peliculas_por_genero.plot(kind="bar")
plt.title('Cantidad de Películas por Género')
plt.xlabel('Género')
plt.ylabel('Número de películas')
plt.xticks(rotation=45, ha="right")
plt.show()
```
